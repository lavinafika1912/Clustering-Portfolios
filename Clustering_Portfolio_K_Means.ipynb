**K-MEANS**

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("/content/Form Andat Clustering_ Sepatu (Responses).xlsx")
```

```python
df.head()
```

```python
df_dataset = df[['Nama', 'tinggi_badan', 'ukuran_sepatu']]
```

```python
df_dataset.head()
```

```python
plt.scatter(df_dataset['tinggi_badan'],df_dataset['ukuran_sepatu'])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df_dataset[['tinggi_badan','ukuran_sepatu']])
y_predicted
```

```python
df_dataset['cluster'] = y_predicted
df_dataset.head()
```

```python
df1 = df_dataset[df_dataset.cluster==0]
df2 = df_dataset[df_dataset.cluster==1]
df3 = df_dataset[df_dataset.cluster==2]

plt.scatter(df1.tinggi_badan,df1['ukuran_sepatu'], color = 'green')
plt.scatter(df2.tinggi_badan,df2['ukuran_sepatu'], color = 'red')
plt.scatter(df3.tinggi_badan,df3['ukuran_sepatu'], color = 'black')

plt.xlabel('tinggi_badan')
plt.ylabel('ukuran_sepatu')
plt.legend()
```

```python
scaler = MinMaxScaler()
df_dataset['ukuran_sepatu'] = scaler.fit_transform(df_dataset[['ukuran_sepatu']])
df_dataset['tinggi_badan'] = scaler.fit_transform(df_dataset[['tinggi_badan']])
df_dataset
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df_dataset[['tinggi_badan','ukuran_sepatu']])
y_predicted
```

```python
df_dataset['cluster'] = y_predicted
df_dataset
```

```python
df1 = df_dataset[df_dataset.cluster==0]
df2 = df_dataset[df_dataset.cluster==1]
df3 = df_dataset[df_dataset.cluster==2]

plt.scatter(df1.tinggi_badan,df1['ukuran_sepatu'], color = 'green')
plt.scatter(df2.tinggi_badan,df2['ukuran_sepatu'], color = 'red')
plt.scatter(df3.tinggi_badan,df3['ukuran_sepatu'], color = 'black')

plt.xlabel('tinggi_badan')
plt.ylabel('ukuran_sepatu')
plt.legend()
```

```python
km.cluster_centers_
```

```python
df1 = df_dataset[df_dataset.cluster==0]
df2 = df_dataset[df_dataset.cluster==1]
df3 = df_dataset[df_dataset.cluster==2]

plt.scatter(df1.tinggi_badan,df1['ukuran_sepatu'], color = 'green')
plt.scatter(df2.tinggi_badan,df2['ukuran_sepatu'], color = 'red')
plt.scatter(df3.tinggi_badan,df3['ukuran_sepatu'], color = 'black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*', label="centroid")

plt.xlabel('tinggi_badan')
plt.ylabel('ukuran_sepatu')
plt.legend()
```

```python
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df_dataset[['tinggi_badan','ukuran_sepatu']])
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
